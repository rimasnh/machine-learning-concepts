```python
# Import libraries.
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Download the California housing data.
# The dataset source: https://www.dcc.fc.up.pt/~ltorgo/Regression/cal_housing.html
housing = fetch_california_housing()
```

```python
# Check the type and the shape of the dataset.
print(type(housing.data))
print(housing.data.shape)
print(type(housing.target))
print(housing.target.shape)
```

```python
# Split the data into train and test. Then split the train into the actual train and the validation.
X_train_full, X_test, y_train_full, y_test = train_test_split(
    housing.data, housing.target)
X_train, X_valid, y_train, y_valid = train_test_split(
    X_train_full, y_train_full)
```

```python
# Scale the data.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)
```

# Part 1: Linear Regression

```python
# Fit the linear regression model.
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression(fit_intercept=True).fit(
    X_train, y_train )
```

Note that the California Housing dataset has the following attributes:
* *MedInc* median income in block group
* *HouseAge* median house age in block group
* *AveRooms* average number of rooms per household
* *AveBedrms* average number of bedrooms per household
* *Population* block group population
* *AveOccup* average number of household members
* *Latitude* block group latitude
* *Longitude* block group longitude

```python
# Examine the linear regression model coefficients.
print("The linear regression model:")
print("Intercept: ", linear_reg.intercept_)
print("The coefficients: ", linear_reg.coef_)
```

```python
# Predict on the train and test sets.
train_preds = linear_reg.predict(X_train)
test_preds = linear_reg.predict(X_test)
```

```python
# Report the train/test error of the linear regression model.
from sklearn.metrics import mean_squared_error
print("Train MSE: ", mean_squared_error(y_train, train_preds))
print("Test MSE: ", mean_squared_error(y_test, test_preds))
```

# Part 2: Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

# Transform the train/test sets using polynomial features.
poly = PolynomialFeatures(2)
poly_X_train = poly.fit_transform(X_train)
poly_X_test = poly.fit_transform(X_test)

# Train the polynomial linear regression model.
poly_linear_reg = LinearRegression().fit(poly_X_train, y_train)

# Predict on the train and test sets.
poly_train_preds = poly_linear_reg.predict(poly_X_train)
poly_test_preds = poly_linear_reg.predict(poly_X_test)

# Report the train/test error of the linear regression model.
from sklearn.metrics import mean_squared_error
print("Train MSE: ", mean_squared_error(y_train, poly_train_preds))
print("Test MSE: ", mean_squared_error(y_test, poly_test_preds))
```

# Part 3: Neural Network Regressor

## Linear Regression in TensorFlow

```python
# Build linear regression in Tensorflow.
from tensorflow import keras
model = keras.models.Sequential([
    keras.layers.Dense(1)
])
model.compile(loss="mean_squared_error", optimizer="adam")
history = model.fit(
    X_train, y_train, epochs=30, verbose=2,
    validation_data=(X_valid, y_valid))
```

```python
# Display the model summary.
model.summary()
```

```python
# Extract the model parameters for the Tensorflow linear regression.
print("TensorFlow linear regression model:")
print(model.layers[0].get_weights())

# Compare against the Scikit-learn model.
print("\nScikit-learn linear regression model:")
print("Intercept: ", linear_reg.intercept_)
print("The coefficients: ", linear_reg.coef_)
```

## Neural Network Regressor With One Hidden Layer

```python
# One hidden layer neural network.
model = keras.models.Sequential([
    keras.layers.Dense(30, activation="relu", input_shape=X_train.shape[1:]),
    keras.layers.Dense(1)
])
model.compile(loss="mean_squared_error", optimizer="adam")
history = model.fit(
    X_train, y_train, epochs=20, verbose=2,
    validation_data=(X_valid, y_valid))
```

```python
# Display the model summary.
model.summary()
```

```python
# Compute the mean squared error on the test set.
mse_test = model.evaluate(X_test, y_test)
print("Mean squared error on the test set: ", mse_test)
```
